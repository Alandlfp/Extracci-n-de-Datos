*Filtro 6: se seleccionan las filas las cuales la columna "life_expectancy" sea mayoy a 77*

```python
filtro6=data[data["life_expectancy"] > 77]
filtro6
```

```python
filtro6.to_csv("Filtro6.csv")
```

*Filtro 7: Se seleccionan las filas que cumplan con "life_expectancy" sea mayor a 77 y "continent" sea "Europe"*

```python
filtro7=data[(data["life_expectancy"] > 77) & (data["continent"]=="Europe")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")
```

*Filtro 8: Se seleccionan las filas que cumplan ""life_expectancy" sea mayor a 77 o que "continent" sea "Europe"*

```python
filtro8=data[(data["life_expectancy"] > 77) | (data["continent"]=="Europe")]
filtro8
```

```python
filtro8.to_csv("Filtro8_csv")
```

*Filtro 9: se seleccionan las filas que no cumpla con filtro8=data "life_expectancy" sea mayor a 77 y que "continent" se "Europe"*

```python
filtro9=data[ ~ (data["life_expectancy"] > 77) & ~ (data["continent"]=="Europe")]
filtro9
```

```python
filtro9.to_csv("Filtro9_csv")
```
